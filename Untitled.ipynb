```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.utils import to_categorical

from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

```python
train=pd.read_csv('train.csv')
train
```

```python
train_image = []

for i in tqdm(range(train.shape[0])):
    imag=image.load_img('images/'+train['image_names'][i],target_size=(28,28,3))
    img = image.img_to_array(imag)
    img = img/255.0
    train_image.append(img)
    
```

```python
y=train['emergency_or_not']
y=to_categorical(y)
x=np.array(train_image)
x_train,x_test, y_train,y_test = train_test_split(x,y,random_state=4,test_size=0.2)
```

```python
x_train.shape
```

```python
from tensorflow.keras import models 
from tensorflow.keras import layers
model = models.Sequential()

model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,3)))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))

model.add(layers.Conv2D(128,(3,3), activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
```

```python
model.add(layers.Flatten())
model.add(layers.Dense(500,activation='relu'))
model.add(layers.Dense(100,activation='relu'))
model.add(layers.Dense(2,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=30,validation_data=(x_test,y_test))
```

```python
test = pd.read_csv('test.csv')
test_img=[]                   
for i in tqdm(range(test.shape[0])):
    imag = image.load_img('images/'+test['image_names'][i],target_size=(28,28,3))
    img = image.img_to_array(imag)
    img = img/255.0
    test_img.append(img)
x_pred = np.array(test_img) 
```

```python
y_pred = model.predict_classes(x_pred)
sample=pd.read_csv('ans.csv')
sample['emergency_or_not']=y_pred
sample.to_csv('sample.csv',header=True,index=False)
```
